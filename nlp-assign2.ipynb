```python
import re
import numpy as np
import nltk
import spacy
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('wordnet')
nltk.download('stopwords')

```

```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model='gpt2')

```

```python
def generate (phrase) :
  set_seed(50)
  text = generator(phrase, max_length=20, num_return_sequences=1)
  text = str(text)
  return text

```

****Preprocessing****

```python
def preprocessing (text) :
  cleaned_text = re.sub(r"\\n\\n" , "" , text)
  cleaned_text = re.sub(r"[^a-zA-Z]" , " " , cleaned_text)
  cleaned_text = re.sub(r"\s+" , " " , cleaned_text )
  normalized_text = cleaned_text.lower()
  tokens = normalized_text.split()
  lemmatizer = WordNetLemmatizer()
  lemmatized_tokens = [lemmatizer.lemmatize(i) for i in tokens]
  #lemmatized_tokens = ' '.join(lemmatized_tokens)
  stop_words = set(stopwords.words("english"))
  filtered_tokens = [token for token in lemmatized_tokens if token not in stop_words]
  #filtered_tokens = ' '.join(filtered_tokens)
  unique_words = set(filtered_tokens)
  unique_words = ' '.join(unique_words)

  return normalized_text , filtered_tokens , unique_words


```

****TF-IDF (From Scratch)****

```python
import math
from collections import Counter

# Calculate Term Frequency in each document in the corpus
def compute_tf(text):
    tf_text = Counter(text)
    total_words = len(text)
    for word in tf_text:
        tf_text[word] = tf_text[word] / total_words
    return tf_text


# Calculate inverse document frequency
def compute_idf(word, corpus):
    N = len(corpus)
    num_docs_containing_term = sum(1 for doc in corpus if word in doc)
    return math.log10((1 + N) / (1 + num_docs_containing_term)) + 1

# Calculate TF_IDF = TF*IDF
def compute_tfidf(corpus):
    tfidf_results = []
    for doc in corpus:
        tfidf_doc = {}
        tf_scores = compute_tf(doc)
        for word in tf_scores:
            tfidf_doc[word] = tf_scores[word] * compute_idf(word, corpus)
        tfidf_results.append(tfidf_doc)
    return tfidf_results

# Normalizing the TF_IDF results : TF_IDF / sum of square TF_IDF
def normalize_tfidf(tfidf_results):
    for doc in tfidf_results:
        sum_of_squares = sum(tfidf ** 2 for tfidf in doc.values())
        for word in doc:
            doc[word] /= math.sqrt(sum_of_squares)
    return tfidf_results



corpus_words = [document.lower().split() for document in corpus]
tfidf_results = compute_tfidf(corpus_words)
normalized_tfidf_results = normalize_tfidf(tfidf_results)


for i, doc in enumerate(normalized_tfidf_results):
    print("Normalized TF-IDF for Document", i + 1)
    for word, score in doc.items():
        print(f"{word}: {score:.4f}")
    print()
```
